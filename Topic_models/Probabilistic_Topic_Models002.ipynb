# **Preprocessing:**

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

import pandas as pd
import numpy as np
from farhad.Farhadcolor import tcolors,bcolors
import warnings
```

```python
df_data = pd.read_csv('data/clean_creditcard_data.csv')
df_data.info()
```

### **Ti_TF for uni_gram and bi_gram:**

```python
%%time
warnings.filterwarnings('ignore')
vectorizer_count_vector = CountVectorizer(min_df=5, max_df=0.9,ngram_range=(1,2))
data_vectorized_vc = vectorizer_count_vector.fit_transform(df_data['new'])

TiTf = TfidfTransformer()
data_vectorized = TiTf.fit_transform(data_vectorized_vc)
```

```python
# Check the Sparsicity: Sparsicity is nothing but the percentage of non-zero datapoints in the document-word matrix,
data_dense = data_vectorized.todense() # Materialize the sparse data
print("Sparsicity: ", ((data_dense > 0).sum()/data_dense.size)*100, "%")
```

# **LDA model:**

```python
from sklearn.decomposition import LatentDirichletAllocation
import pickle
```

```python
%%time
warnings.filterwarnings('ignore')

number_of_classfication = 11

LDA = LatentDirichletAllocation(n_components=number_of_classfication, learning_decay=0.9, learning_method='batch', batch_size=50, verbose=1, random_state=32)
result_lda = LDA.fit_transform(data_vectorized)
label_sk = np.argmax(result_lda,axis=1)

print(' ')
```

```python
def dominant_topic_in_each_document(best_lda_model,lda_output, n_topics=11):
    """ 
    info:
    
    -----------------------------
    input:
    
    -----------------------------
    ouput:
    
    ------------------------------- 
    """
    #lda_output = best_lda_model.transform(data_vectorized)

    # column names
    topicnames = ["Topic" + str(i) for i in range(n_topics)]

    docnames = ["Doc" + str(i) for i in range(len(lda_output))]

    # Make the pandas dataframe
    df_document_topic = pd.DataFrame(np.round(lda_output, 2), columns=topicnames, index=docnames)

    # Get dominant topic for each document
    dominant_topic = np.argmax(df_document_topic.values, axis=1)
    df_document_topic['dominant_topic'] = dominant_topic

    # Styling
    def color_green(val):
        color = 'green' if val > .1 else 'black'
        return 'color: {col}'.format(col=color)

    def make_bold(val):
        weight = 700 if val > .1 else 400
        return 'font-weight: {weight}'.format(weight=weight)

    # Apply Style
    show_document_topics = df_document_topic.head(15).style.applymap(color_green).applymap(make_bold)
    df_document_topics = df_document_topic.style.applymap(color_green).applymap(make_bold)
    
    return show_document_topics, df_document_topic
```

```python
def Extract_topic_words(model,vectorizer,top_n):
    """ 
    info:
    
    -----------------------------
    input:
    
    -----------------------------
    ouput:
    
    ------------------------------- 
    """
    
    topic_words = {}
    word_idx = []
    for idx, topic in enumerate(model.components_):
        word_idx.append(idx)
        topic_words[str(idx)] = [vectorizer.get_feature_names()[i] for i in topic.argsort()[:-top_n - 1:-1]]
       
    for idx in word_idx :
        
        print('Topic: %d' % idx)
        print(topic_words[str(idx)])
        print(' ')
        print(bcolors.BLUE,'                                                    ',bcolors.ENDC)
    
    
    return topic_words
def Topic_keywords(best_lda_model,vectorizer,n_topics=11):
    """ 
    info:
    
    -----------------------------
    input:
    
    -----------------------------
    ouput:
    
    ------------------------------- 
    """
    
    # Topic-Keyword Matrix
    topicnames = ["Topic" + str(i) for i in range(n_topics)]
    df_topic_keywords = pd.DataFrame(best_lda_model.components_)
    # Assign Column and Index
    df_topic_keywords.columns = vectorizer.get_feature_names()
    df_topic_keywords.index = topicnames
    
    return df_topic_keywords
```

## 1. How to GridSearch the best LDA model?

```python
from sklearn.model_selection import GridSearchCV
from sklearn.externals import joblib
```

```python
%%time
search_params = {'n_components': [4,5,6,7,8,9,10,11,12], 'learning_decay': [.5, .7, .9]}

lda = LatentDirichletAllocation(batch_size=128, doc_topic_prior=None,
             evaluate_every=-1, learning_decay=0.7, learning_method=None,
             learning_offset=10.0, max_doc_update_iter=100, max_iter=10,
             mean_change_tol=0.001, n_components=10, n_jobs=1,
             n_topics=None, perp_tol=0.1, random_state=None,
             topic_word_prior=None, total_samples=1000000.0, verbose=0)

# Init Grid Search Class GridSearchCV(lda, param_grid=search_params)
model = GridSearchCV(cv=None, error_score='raise',
       estimator=lda,      
       fit_params=None, iid=True, n_jobs=1,
       param_grid= search_params,
       pre_dispatch='2*n_jobs', refit=True, return_train_score='warn',
       scoring=None, verbose=0)
# Do the Grid Search
model.fit(data_vectorized)
```

```python
filename = 'topic_model_n11.sav'
joblib.dump(model, filename)
#loaded_model = joblib.load(filename)
```

```python
print(bcolors.GREEN,'                                                               ',bcolors.ENDC)

# Best Model
best_lda_model = model.best_estimator_

# Model Parameters
print("Best Model's Params: ", model.best_params_)

# Log Likelihood Score
print("Best Log Likelihood Score: ", model.best_score_)

# Perplexity
print("Model Perplexity: ", best_lda_model.perplexity(data_vectorized))

print(bcolors.GREEN,'                                                               ',bcolors.ENDC)
```
