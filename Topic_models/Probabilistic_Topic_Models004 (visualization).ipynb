```python
from sklearn.cluster import KMeans
from sklearn.decomposition import TruncatedSVD
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
result_lda2 = []

for num in result_lda.keys():
    result_lda2.append(result_lda[num])
```

```python
result_lda2[1]
```

```python
# Construct the k-means clusters
clusters = KMeans(n_clusters=11, random_state=100).fit_predict(result_lda2)

# Build the Singular Value Decomposition(SVD) model
svd_model = TruncatedSVD(n_components=2)  # 2 components
lda_output_svd = svd_model.fit_transform(result_lda2)

# X and Y axes of the plot using SVD decomposition
x = lda_output_svd[:, 0]
y = lda_output_svd[:, 1]

# Weights for the 15 columns of lda_output, for each component
print("Component's weights: \n", np.round(svd_model.components_, 2))

# Percentage of total information in 'lda_output' explained by the two components
print("Perc of Variance Explained: \n", np.round(svd_model.explained_variance_ratio_, 2))
```

```python
# Plot
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(12, 12))
pp = plt.scatter(x, y, c=clusters, cmap='viridis',alpha=0.6, edgecolors='none', s=30, label=clusters)
plt.colorbar()
plt.legend()
plt.xlabel('Component 2')
plt.xlabel('Component 1')
plt.title("Segregation of Topic Clusters", )
#plt.savefig('Picture_result/hearmaps_topic_worsd5.png')
plt.show()
```

```python
lda_output_svd[1]
```

```python
part_dic_x = {}
part_dic_y = {}
for num in range(11):
    part_dic_x[str(num)] = []
    part_dic_y[str(num)] = []
for num,classi in enumerate(clusters):
    part_dic_x[str(classi)].append(x[num])
    part_dic_y[str(classi)].append(y[num])
```

```python
# Plot
labels = ['Topic_'+str(i) for i in clusters]
colors = ["red", "green", "blue",'#006600','#0088CC','#DD99FF','#660022','#B35900','#666600','#FF4019','#804000']

plt.figure(figsize=(12, 12))

for num  in range(len(part_dic_x)):
    plt.scatter(part_dic_x[str(num)], part_dic_y[str(num)], c = colors[clusters[num]], alpha=0.6, edgecolors='none', s=30, 
                label=labels[clusters[num]])
    
#plt.colorbar()
plt.legend()
plt.xlabel('Component 2')
plt.xlabel('Component 1')
plt.title("Segregation of Topic Clusters", )
#plt.savefig('Picture_result/hearmaps_topic_worsd6.png')
```

```python
import matplotlib as mpl
mpl.use('Agg')  # No display
import matplotlib.pyplot as plt
import io

import tensorboard as tb
import tensorflow as tf
```

```python
%load_ext tensorboard.notebook
```

```python

# Plot
def get_plot_buf():
    plt.figure(figsize=(12, 12))
    plt.style.use('seaborn-darkgrid')

    pp = plt.scatter(x, y, c=clusters, cmap='viridis',alpha=0.6, edgecolors='none', s=30, label=clusters)
    plt.colorbar()
    plt.legend()
    plt.xlabel('Component 2')
    plt.xlabel('Component 1')
    plt.title("Segregation of Topic Clusters", )
    #plt.savefig('Picture_result/hearmaps_topic_worsd5.png')
    plt.show()
    buf = io.BytesIO()
    plt.savefig(buf, format='png')
    plt.close()

    buf.seek(0)
    return buf
```

```python
session = tf.Session()

plot_buf_ph = tf.placeholder(tf.string)
image = tf.image.decode_png(plot_buf_ph, channels=4)
image = tf.expand_dims(image, 0)  # make it batched
plot_image_summary = tf.summary.image('some_name', image, max_outputs=1)

# later, to make the plot:
plot_buf = get_plot_buf()
plot_image_summary_ = session.run(
        plot_image_summary,
        feed_dict={plot_buf_ph: plot_buf.getvalue()})
#summary_writer.add_summary(plot_image_summary_, global_step=iteration)
```
